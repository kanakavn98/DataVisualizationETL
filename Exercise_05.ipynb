```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/content/student (1).csv')
# to show the first five rows
df.head()
# to show last 5 rows
#df.tail()
```

```python
#create histogram of maths score
#old fashion way
math_bins=df['math_score'].value_counts(bins=10)

math_bins
```

```python
math_bins.plot(kind='bar')
plt.xticks(rotation=20)
plt.show()
# incorrect to draw bar graph because the data is not contionous and moreover it doenot start with zero the minimum value of maths score
```

```python
# now the perfect histogram to plot the data
# kde is used for drawing more scatter line as shown below
sns.histplot(data=df ,x='math_score',color='#e07a5f',bins=20,kde=True)
plt.show()
```

```python
# creating box plot for maths score
sns.boxplot(data=df,x='math_score',color='red')
plt.show()
```

```python
# drawing the hist plot for writing score
sns.histplot(data=df ,x='writing_score',color='pink',bins=20)
plt.show()
```

```python
sns.boxplot(data=df,x='writing_score',color='green')
plt.show()
```

```python
# creating box plot for ethnicity variable
# for categorical variable we can also use boxplot by using catplot function
# combine one numeric variable and one categorical varaiable together to create groups of boxplot
sns.catplot(data=df,x='ethnicity',y='math_score', kind = 'box')
plt.show()
```

```python
# Showing the boxplot of reading score across lunch choice
# for categorical variable we can also use boxplot by using catplot function
# combine one numeric variable and one categorical varaiable together to create groups of boxplot
sns.catplot(data=df,x='lunch',y='reading_score', kind = 'box')
plt.title('Reading Score distribution across lunch plan')
plt.show()
```

```python
#histogram (smoother version of histogram ,kde= True)
sns.kdeplot(data=df, x= 'math_score',fill=True, color='orange',bw_adjust=0.2)
plt.show()
```

```python
#how maths score is distributed across gender ? kde plot
sns.kdeplot(data=df,x='math_score',hue='gender',common_norm=False, palette={'female':'pink','male':'blue'})

plt.show()
```

```python
#how maths score is distributed across ethnicity ? kde plot
# we use Common_norm to seggragate the value of peak
sns.kdeplot(data=df,x='math_score',hue='ethnicity',common_norm=False,fill= True, alpha=0.1,
             palette={'group A':'pink','group B':'blue','group C':'yellow','group D' : 'red','group E':'orange'})

plt.show()
# below graph show group E has goof marks in maths
```

```python
# how plot scatter plot
# we will explore distribution of maths and reading score
sns.scatterplot(data=df, x='math_score',y='reading_score')
plt.title(" reading score across maths score x")
plt.show()
# conclusion is that the result is linear means positive
```
