```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df= pd.read_csv('/content/student (1).csv')
df.head()
```

```python
# how to visualize the maths score distribution across the parental_education
sns.catplot(data=df,x='parental_education',y='math_score', kind = 'box')
plt.xticks(rotation=20)
plt.show()
```

```python
#we can also use below code for both distribution
# we need to separate the data in subgroup , then we need to make kde also separate so we use common_norm as FALSE so that we can make better
#visualization
sns.kdeplot(data=df,x='math_score',hue='parental_education',common_norm=False)
plt.show()
```

***The above graph shows Master parents childrens tends to have good maths score ***

```python
# we can also use histrogram for the above question
# when we use kde = true then we show the curves too
# when we use set1 we are moving to darker shades ...and set 3 lighter color
sns.histplot(data=df,x='math_score',hue='parental_education',kde=True, palette='Set1')
plt.show()
```

```python
#Task 2
#Create a visualization showing the relationship between reading and writing score and show they are distributed across gender
#using learning resource in canvas nd go to link  marker , to chnge marker
sns.scatterplot(data=df, x='reading_score',y='writing_score',hue='gender',marker="^",s=150,palette={'female':'red','male':'green'})
plt.show()
```

```python
#Task 3
#Create a visualization showing the relationship between reading and maths_score and show they are distributed across gender
sns.scatterplot(data=df, x='reading_score',y='math_score',hue='gender')
plt.show()
```

```python
# to develop the reggression line
sns.jointplot(data=df, x='math_score', y='reading_score',kind='reg',color='orange',scatter_kws={'alpha':0.3})
plt.show()
```

```python
from seaborn._core.properties import LineStyle
# visualizing the gold price chnge over time
plt.figure(figsize=(15,7))
# plt.plot_date():to identify the date column nd automatically put the x axis as date
#mfc: to just chnge the colour of marker
#mec: to adjust edge color of the marker
plt.plot_date(x=df_gold['Date'],y=df_gold['Gold'],linestyle='--',c='#c9a227',marker="X",markersize=10,mfc='red',mec='black')
# below code is used to fix rotation automatically
plt.gcf().autofmt_xdate()
#plt.xticks(rotation=40)
plt.show()
```

```python
df_gold= pd.read_csv('/content/annual_gold.csv')
df_gold['Date'].str[:7]
df_gold.head()
```

```python
#visualizing platinum using seaborn library
plt.figure(figsize=(15,7))
sns.lineplot(data=df_gold,x='Date',y='Platinum',marker='*',markersize=12, linestyle='--')
# below code is used to fix rotation automatically
plt.gcf().autofmt_xdate()
plt.grid(True)
plt.show()
```

```python
#create one graph thta contain both gold and platinum together
plt.figure(figsize=(15,7))
# plot the gold
sns.lineplot(data=df_gold,x='Date',y='Gold',marker="p",linestyle='--',markersize=10,color='orange',mfc='red',mec='black')

# plot the platinum
sns.lineplot(data=df_gold,x='Date',y='Platinum',marker="p",linestyle='--',markersize=10,color='green',mfc='red',mec='black')
plt.gcf().autofmt_xdate()
plt.grid(True)
plt.show()
```
