```python
# visulaize the value_counts results using bar chart
reading_score_bins=df['reading_score'].value_counts(bins=6) # storing the value
plt.figure(figsize=(8,4))
reading_score_bins.plot(kind='bar',color='purple')
plt.title('Distributions of Reading scores in Class')

plt.xlabel('bins or distributions ')
plt.ylabel('no falls under bins')
plt.xticks(rotation=20)
plt.show()



```

```python
# visulaize the value_counts results using bar chart
gender=df['gender'].value_counts() # storing the value
plt.figure(figsize=(8,4))
gender.plot(kind='bar',color='orange')
plt.title('Distributions of gender in Class')

plt.xlabel('x ')
plt.ylabel('y')
plt.xticks(rotation=0)

plt.show()
```

```python
# to shows gender distribution across parental_education
# define 2d contigency table
# contigency_table= pd.crosstab(rows, column)
contigency_table= pd.crosstab(df['parental_education'],df['gender'])
contigency_table
```

```python
# visulaize the  above table
# cmap is used to high darker blue nd lower had really light blue , annot is set true for making the value as displayed on maps
# fmt='d' means data type to proper number ,linewidths=0.5 separates the two enteries of the box
sns.heatmap(contigency_table, cmap='Blues',annot= True, fmt='d', linewidths=0.5)

plt.show()
```

```python
# changeing table into  percentage value
percentage_table =contigency_table/1000*100
percentage_table
# visulaize the  above table
# cmap is used to high darker blue nd lower had really light blue , annot is set true for making the value as displayed on maps
# fmt='.2f' means showing float value at 2 decimal places ,linewidths=0.5 separates the two enteries of the box
sns.heatmap(percentage_table, cmap='coolwarm',annot= True, fmt='.2f', linewidths=0.5)

plt.show()
```

```python
# Create the bar chart showing gender distribution over test_preparation course
plt.figure(figsize=(8,4))
# countplot each label and have the value quantity of it in the label, using hue will distrbute among gender
# for manually setting the color of female as pink we use {} under palette
sns.countplot(data=df,x='parental_education', hue='gender',
              palette={'female':'pink', 'male':'skyblue'})
plt.xticks(rotation=20)
plt.show()
```
