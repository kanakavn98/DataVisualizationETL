```python
import matplotlib.pyplot as plt


```

```python
med=[35,15,20,40,25]
time=['Monday','Tuesday','Wednesday','Thursday','Friday']
```

```python
plt.figure (figsize=(8,4))
plt.plot(time,med,color='#DE6B48',label='Dosage')
plt.fill_between(time,med, color='#F05D0E', alpha= 0.1)
plt.title('Dosage of medication over Time')
plt.grid(True)
plt.xlabel('Time(day)')
plt.ylabel('Dosage')
plt.legend()
plt.show()
```
