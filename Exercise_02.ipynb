```python
# create variable x
x= np.linspace(0,20,40)
print (x)
# calculates length
len(x)
```

```python
#create y based on x
y=np.sin(x)
y
```

```python
#create a figure
plt.figure(figsize=(8,4))

# plot the sine wave
plt.plot(x,y, color='#AE7CC2', label='sine(x)')
plt.title('Sine Wave')
# grid is used to define lines within graph
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Amplitude')
# legend is used to show label
plt.legend()
plt.show()
```

```python
#Exercise 1
c= np.linspace(0,100,50)
print (c)
# calculates length
len(c)

# creating y ie h(c) function using c values
y=np.square(c)
y

```

```python
#create a figure
plt.figure(figsize=(8,4))

# plot the function
plt.plot(c,y, color='#AE7CC2', label='c square')
plt.title('Class Work 1- A simple plot')
plt.fill_between(c,y, color='#F05D0E', alpha= 0.1)
# grid is used to define lines within graph
plt.grid(True)
plt.xlabel('c')
plt.ylabel('h(c)')
# legend is used to show label
plt.legend()

plt.show()
```

```python
#Exercise 2
c= np.linspace(0,100,50)
print (c)
# calculates length
len(c)

# creating y ie h(c) function using c values
y=np.log(c)
y
```

```python
#create a figure
plt.figure(figsize=(8,4))

# plot the function
plt.plot(c,y, color='#AE7CC2', label='log function')
plt.title('Class Work 2-Log Function')
plt.fill_between(c,y, color='#F05D0E', alpha= 0.1)
# grid is used to define lines within graph
plt.grid(True)
plt.xlabel('log(c)')
plt.ylabel('h(c)=log(c)')
# legend is used to show label
plt.legend()

plt.show()
```

```python
# using seaborn
import seaborn as sns

```

```python
med=[35,15,20,40,25]
time=['Monday','Tuesday','Wednesday','Thursday','Friday']
```

```python
plt.figure (figsize=(8,4))
sns.lineplot(x=time, y=med,marker='o', markersize= 10, linestyle='--', color='purple')
# marker represents dots , markersize tells the size of dots
plt.fill_between(time,med, color='#9f86c0', alpha= 0.4)
plt.title('Dosage of medication over Time')
plt.grid(True)
plt.show()
```

```python
c= np.linspace(0,100,50)
print (c)
# calculates length
len(c)

# creating y ie h(c) function using c values
y=np.tan(c)
y
```

```python
plt.figure (figsize=(12,4))
sns.lineplot(x=c,y=y,marker='o', markersize=6, linestyle='--', color='red')
# marker represents dots , markersize tells the size of dots
plt.fill_between(c,y, color='orange', alpha= 0.6)
plt.title('Tan Function')
plt.xlabel('c')
plt.ylabel('h(c)=tan(c)')
plt.grid(True)
plt.show()
```
