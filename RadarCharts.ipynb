```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Creating Radar chart for wonder woman
# Step 1: define wonder woman attributees and values
attributes= ['Power','Speed', 'Strength', 'Intelligence','Flight']
values= [90, 80, 75, 90, 75]
```

```python
#Step 2 : Define angles (start,stop,number of attributes)
angles= np.linspace(0,2*np.pi, 5,endpoint=False).tolist() # gives the array but we want list so
angles
angles += angles[:1] # angles[:1] gives the first value of list here its 0.0 , so it will add 0.0 in list
angles
```

```python
# : Define values (start,stop,number of attributes)
values += values[:1]
values
```

```python
# Step 3 : Initialize the radar plot
plt.figure(figsize=(6,6))
ax = plt.subplot(111,polar= True) # polar is used to indicate that the figure is circular

# draw xticks
plt.xticks(angles[:-1], attributes) #while xticks must be equal to no of attributes but

# draw yticks
ax.set_rlabel_position(0)
plt.yticks([20,40,60,80,100],['20', '40', '60','80','100'], color= 'lightgrey', size =10)
plt.ylim(0,100) # because our ylimit is from 0 to 100 ie our data boundry is 0 to 100

# plot data
ax.plot(angles, values, linewidth=1, linestyle ='solid', color='blue') #plot and fill required numberof attributes + 1 as we did for angles+1
ax.fill(angles, values, 'b', alpha=0.3, facecolor='skyblue')

plt.show()
```

```python
hero_data = pd.read_csv('/content/hero_profile.csv')
# to show the first five rows
hero_data.head()
```

```python
def my_radar_chart(hero_data):
  labels= np.array(['power','speed', 'defense', 'strength','flight'])
  num_vars= len(labels)

  # define angles
  angles= np.linspace(0,2*np.pi, num_vars, endpoint=False).tolist()
  angles += angles[:1]

  # initialize radar plot
  fig,ax= plt.subplots(figsize=(6,6), subplot_kw= dict(polar= True))

  # define the colors
  colors= ['lightgreen','red','steelblue']

  #plot each hero data with distinct color
  for index,(row,color) in enumerate (zip(hero_data.iterrows(),colors)):
    hero_name = row[1]['hero']
    stats= row[1].drop('hero').values
    stats = np.concatenate((stats,[stats[0]]))

    ax.plot(angles, stats, linewidth= 2, label= hero_name, color= color)
    ax.fill(angles, stats, alpha=0.1, color= color)
  ax.set_xticks(angles[:-1])
  ax.set_xticklabels(labels)
  ax.set_yticklabels([])
  plt.legend(loc= 'upper right', bbox_to_anchor=(1.1,1.1))

  return fig, ax
```

```python
fig, ax = my_radar_chart(hero_data)
```
