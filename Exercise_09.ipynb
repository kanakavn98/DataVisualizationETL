```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
# visualize a line in 3d
#y=3x+7

x_line = np.linspace(0, 10, 100)# to define the lower an dupper boundry; 100 means visualizing 100 data point bwteen 0 to 10
y_line= 3* x_line + 7
# when we r assuming zline as zero thn we need to have same data points as x in z_line
z_line= np.zeros_like(x_line)# zeros means matrix of zero with same shape as x_line


# set up the figure and 3d axis
fig= plt.figure(figsize=(8,6))
ax= fig.add_subplot(111,projection='3d')# 111 represents that the image is to be 3d

#plot the line
ax.plot(x_line,y_line,z_line,color='red',label='3*x+7',linewidth=3)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
plt.show()
```

```python
# visualize a line in 3d
#z= 2*x +y +5
# z= f(x,y)
# defining the function
def f(x,y):
  return 2*x +y + 5

# defining the x and y axis
x = np.linspace(-1, 5, 10)# to define the lower and upper boundry; 100 means visualizing 100 data point bwteen 0 to 10
y = np.linspace(-1, 5, 10)

# we need to draw a hyperplane then we need to combine x and y to form matrix so that it will became base for 3d
X,Y =np.meshgrid(x,y)
Z = f(X,Y)

# set up the figure and 3d axis
fig= plt.figure(figsize=(8,6))
ax= fig.add_subplot(111,projection='3d')# 111 represents that the image is to be 3d

#plot the line
ax.plot_wireframe(X,Y,Z, color='green',label='z= 2*x +y +5',linewidth=2)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.legend()
plt.show()
```

```python
# visualize a line in 3d
#z= 3*x*x + 2*y*y --- curved hyperplane and nonlinear function
# z= f(x,y)
# defining the function

def f(x,y):
  return 3*x*x + 2*y*y

# defining the x and y axis
x = np.linspace(-1, 5, 10)# to define the lower and upper boundry; 100 means visualizing 100 data point bwteen 0 to 10
y = np.linspace(-1, 5, 10)

# we need to draw a hyperplane then we need to combine x and y to form matrix so that it will became base for 3d
X,Y =np.meshgrid(x,y)
Z = f(X,Y)

# set up the figure and 3d axis
fig= plt.figure(figsize=(8,6))
ax= fig.add_subplot(111,projection='3d')# 111 represents that the image is to be 3d

#plot the line
#ax.plot_wireframe(X,Y,Z, color='green',label='z= 3*x*x + 2*y*y',linewidth=2)
# we can aslo use plot surface :
ax.plot_surface(X,Y,Z)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.legend()
plt.show()
```

```python
# visualize a line in 3d
#z= cos.sqrt(x^2 +y^2)
# z= f(x,y)
# defining the function

def f(x,y):
  return np.cos(np.sqrt((x**2)+(y**2)))

# defining the x and y axis
x = np.linspace(-1, 5, 10)# to define the lower and upper boundry; 100 means visualizing 100 data point bwteen 0 to 10
y = np.linspace(-1, 5, 10)

# we need to draw a hyperplane then we need to combine x and y to form matrix so that it will became base for 3d
X,Y =np.meshgrid(x,y)
Z = f(X,Y)

# set up the figure and 3d axis
fig= plt.figure(figsize=(8,6))
ax= fig.add_subplot(111,projection='3d')# 111 represents that the image is to be 3d

#plot the line
ax.plot_wireframe(X,Y,Z, color='green',label='z= cos.sqrt(x^2 +y^2)',linewidth=1)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.legend()
plt.show()
```

```python
# visulaize function with changing the boundry of linespace
#z= log(x^2 +y^2)
# z= f(x,y)
# defining the function

def f(x,y):
  return np.log((x**2)+(y**2))

# defining the x and y axis
x = np.linspace(0.1, 10, 50)# to define the lower and upper boundry; 100 means visualizing 100 data point bwteen 0 to 10
y = np.linspace(0.1, 10, 50)

# we need to draw a hyperplane then we need to combine x and y to form matrix so that it will became base for 3d
X,Y =np.meshgrid(x,y)
Z = f(X,Y)

# set up the figure and 3d axis
fig= plt.figure(figsize=(8,6))
ax= fig.add_subplot(111,projection='3d')# 111 represents that the image is to be 3d

#plot the line
#ax.plot_wireframe(X,Y,Z, color='green',label='z= log(x^2 +y^2)',linewidth=4)
ax.plot_surface(X,Y,Z,cmap='hsv')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.legend()
plt.show()
```

```python
# visualize bar chart in 3d space

# generation data
sales=[10,20,15,24]
quarter=['Q1','Q2','Q3','Q4']


#create an arrray with the position of each bar along the x-axis
x_pos=np.arange(4)

# set up the figure and 3d axis
fig= plt.figure(figsize=(12,12))
ax= fig.add_subplot(111,projection='3d')# 111 represents that the image is to be 3d

# plot
ax.bar(x_pos,sales,zdir='y',alpha=0.8)
ax.set_xticks(x_pos)
ax.set_xticklabels(quarter)

ax.set_ylabel('Quarter')
ax.set_zlabel('Sales')

fig.show()
```
