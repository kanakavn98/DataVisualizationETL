```python
import pandas as pd
from wordcloud import WordCloud,STOPWORDS
import numpy as np ## for data operations
import seaborn as sns
import matplotlib.pyplot as plt
import PIL.Image # helps in loading images and visualization
```

```python
# let python read the text file and store the poem "She walks in beauty"
# r means reading and if we w means writing the txt
text = open("/content/she_walks_in_beauty.txt", "r").read()
print(text)
```

```python
# transform into wordcloud but before that we will use Stop words
print(STOPWORDS)
```

```python
#visualizing our text
plt.figure(figsize=(8,4))
# we are generating to create the wordcloud
#WordCloud(values to chnge color plattee or background color).generate(text)
# and  can use colourmatplot lib from learning resource for chnge in color
wc=WordCloud(colormap='cool',background_color='lightgrey',
             stopwords= STOPWORDS).generate(text)
plt.imshow(wc)
plt.axis('off')# to remove all of the axis
plt.show()
```

```python
# read the png file
# as we know compt used to store in number format so we will extract those number
#each matrix represents the one point of the image we dont need to visualize but for storage we need
# the matrix of image
pic_heart = np.array(PIL.Image.open('/content/heart.png'))
#pic_heart
```

```python
# to visulaize the text in  image of heart

plt.figure(figsize=(4,4))
# we use mask in for putting text into heart image
wc=WordCloud(colormap='cool',background_color='black',
             stopwords= STOPWORDS, mask = pic_heart,contour_color='white',contour_width=6).generate(text)
plt.imshow(wc)
plt.axis('off')# to remove all of the axis
plt.show()
```
