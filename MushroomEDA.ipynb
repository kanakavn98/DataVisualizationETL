```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df= pd.read_csv('/content/mushroom.csv')
df.head()
```

```python
c=df['class'].value_counts()
```

```python
v=sns.countplot(data=df,x=c)
plt.show()
```

```python
# or we can use pie chart since variable are less thn 5
plt.figure(figsize=(2,2))
plt.pie(c,autopct='%1.1f%%')
plt.show()
```

```python
#Create one graph to display the relationship
#between stem height and stem width,#and
#how edible and poisonous mushrooms are distributed among them.

sns.scatterplot(data=df,x='stem-height', y='stem-width',hue='class')
plt.show()

```

```python
#Create a visualization to display the distribution of stem height.
sns.histplot(data=df ,x='stem-height',color='#e07a5f',bins=20,kde=True)
plt.show()
```

```python
# we can also use kde plot for this
sns.kdeplot(data=df, x= 'stem-height',fill=True, color='orange',bw_adjust=0.2)
plt.show()
```

```python
#Create a visualization to display the median of stem height over variable 'class'.
#Explain the results based on your observation
#
sns.boxplot(data=df,x='class',y='stem-height',color='orange')
plt.show()
```
