```python
# Task to get score  75 of maths if marks are below 70 ,define function

def grade_helper(score):
  if score< 70:
    return 75
  else:
    return score

a= 99
grade_helper(a)
```

```python
df['adjusted_score']= df['Math_score'].apply(grade_helper) # apply helps us to define function and get it inserted into it
df.head()
```

```python
# map the maths to lettered grade A,B,C,D,F
def lettered_grade(score):
  if score >= 90:
    return 'A'
  elif score >= 80:
    return 'B'
  elif score >= 70:
    return 'C'
  elif score >= 60:
    return 'D'
  else:
    return 'F'

a= 98
lettered_grade(a)
```

```python
df['adjusted_Lettered_score']= df['Math_score'].apply(lettered_grade) # apply helps us to define function and get it inserted into it
df.head()
```
